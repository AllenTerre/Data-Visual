```python
import pandas as pd
import numpy as np
import plotly
```

```python
# read the data at first 
data = pd.read_csv("Flights dataset.csv")
data.head(10)
```

```python
# pick up the data we need to use. 
columns_names = data.columns
columns_names
```

```python
data.isna().sum()
```

```python
# As we can see that, there are too many null value appearing in the last of columns. we need to
# drop them all in case of affecting the accuracy of data. 
drop_data = data.drop(['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY',
       'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY', 'Unnamed: 35', 'CANCELLATION_CODE'], axis = 1)

drop_data.head(10)
```

```python
# Now, even if we delete all the null data based on drop_data, we can also keep most of data,
# which will not generate influence of overall accuracy. 
clean_data = drop_data.dropna()
clean_data.isna().sum()
```

```python
# Now, we want to know which state in US has the longest air time. 

import plotly.graph_objects as go
import dash
import dash_core_components as dcc
import dash_html_components as html

group_air_length = clean_data[['ORIGIN_STATE_ABR', 'AIR_TIME']].groupby("ORIGIN_STATE_ABR").apply(np.sum, axis=0)
states_list = group_air_length['ORIGIN_STATE_ABR'].index.tolist()
group_air_length_list = group_air_length['AIR_TIME'].tolist()
group_air_length_list = np.round(np.divide(group_air_length_list, 1000),2)

group_delay = clean_data[['ORIGIN_STATE_ABR', 'ARR_DELAY']].groupby('ORIGIN_STATE_ABR').apply(np.mean, axis=0)
group_delay_list = np.round(group_delay['ARR_DELAY'].tolist(),2)
group_distance = clean_data[['ORIGIN_STATE_ABR', 'DISTANCE']].groupby('ORIGIN_STATE_ABR').apply(np.sum, axis=0)
group_distance_list = group_distance['DISTANCE'].tolist()

array_data = np.column_stack((states_list, group_air_length_list, group_delay_list, group_distance_list))
new_data = pd.DataFrame(data=array_data, columns=['States in US', 'Air length', 'Mean delay time',
                                                 'Air Distance'])

# add text in the map 

new_data['text'] = ' States names : ' + new_data['States in US'] + '<br>' + \
                    ' Air lenght: ' + new_data['Air length'] + 'k hours' + '<br>' + \
                    ' Mean delay time: ' + new_data['Mean delay time'] + 'hours' + '<br>' + \
                    ' Air Distance: ' + new_data['Air Distance'] + 'km'

```

```python

fig = go.Figure(data=go.Choropleth(locations=states_list,
                z=group_air_length_list, 
                text=new_data['text'],
                locationmode = 'USA-states',
                colorscale = 'Reds',
                colorbar_title = "k hours"))

fig.update_layout(
    title_text = '2019 Air time in different states',
    geo_scope='usa', # limite map scope to USA
)


app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app.run_server(debug=True, use_reloader=False)  # Turn off reloader if inside Jupyter
```
